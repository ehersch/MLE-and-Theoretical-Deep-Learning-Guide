# Secretary Matching Problem

Setup: assume we are interviewing $N$ total secretaries for a position and once we interview a candidate, we can either accept or reject immediately (no going back). This is a famous problem from the 1950s and the question is what is the optimal strategy for picking a candidate? How much do we explore first before making our decision?

This notebook is inspired by *Algorithms to Live by*

First, assume every secretary has a positive integer score associated to it to represent its value. The sore is monotonic and uniformly distributed across the population.

```julia
function get_scores(N::Int64)
  scores = collect(1:N)
  # Now shuffle the scores to make a random interview ordering
  scores = shuffle(scores)
  return scores
end
```

```julia
function simulate(scores::Vector{Int64}, policy::Int64)
  best_score = 0
  for i in 1:policy
    best_score = max(best_score, scores[i])
  end
  for i in policy+1:length(scores)
    if scores[i] > best_score
      return scores[i]==N
    end
  end
  return scores[end]==N
end
```

```julia
N = 100 # assume population is 100
S = 50000 # number of samples to draw

EV = zeros(Float64, N)

for _ in 1:S
  scores = get_scores(N)
  for i in 1:N-1
    score = simulate(scores, i)
    EV[i] += score
  end
end

EV = EV / S
nothing
# for i in 1:N-1 # can't just reject all candidates
#   # perform N samples to get EV
#   cur_scores = 0
#   for _ in 1:S
#     scores = get_scores(N)
#     score = simulate(scores, i)
#     cur_scores += score
#   end
#   #println(cur_scores)
#   cur_EV = cur_scores / S
#   EV[i] = cur_EV
# end
```

```julia
using Random, Statistics

function simulate_secretary(N::Int, policy::Int)
    scores = shuffle(1:N)
    best_score = 0
    # observe first 'policy' candidates
    for i in 1:policy
        best_score = max(best_score, scores[i])
    end
    # then take first better than all before
    for i in policy+1:N
        if scores[i] > best_score
            return scores[i] == N
        end
    end
    return scores[end] == N
end

function monte_carlo(N::Int, trials::Int)
    results = []
    for policy in 1:N
        success = mean(simulate_secretary(N, policy) for _ in 1:trials)
        push!(results, success)
    end
    return results
end

N = 100
trials = 100000
EV = monte_carlo(N, trials)
best_index = argmax(EV)
println("Best policy ≈ ", best_index, ", Success Rate = ", EV[best_index])
```
