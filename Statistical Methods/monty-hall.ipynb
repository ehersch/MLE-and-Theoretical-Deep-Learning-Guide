# Solving the Monty Hall Problem with Monte-Carlo simulation

__Setup:__ You're a contestant on a game show presented with 3 doors. Goats are behind 2 and a car is behind 1. You select one door and the host opens one of the doors with a goat. You have an option: you can keep your current door or switch to the other one. What do you do?

Supposedly, your odds of winning are much higher if you decide to switch. Let's simulate it.

```python
import math
import random
```

__Baseline:__ This is simple, just choose a door and see what its success rate is... should be $\frac{1}{3}$

```python
def baseline(N):
  success = 0

  for i in range(N):
    choice = random.randint(1, 3) # choose door 1, 2, 3
    car = random.randint(1, 3) # car is in one of the 3

    if choice == car:
      success += 1
  
  return success / N

baseline(100000)
```

```python
def switch(N):
    success = 0

    for _ in range(N):
        options = [1, 2, 3]

        choice = random.choice(options)  # your initial pick
        car = random.choice(options)     # where the car is

        # Host opens a goat door (not your choice, not the car)
        remaining = [d for d in options if d != choice and d != car]
        goat = random.choice(remaining)

        # Switch to the only remaining unopened door
        switch_choice = [d for d in options if d != choice and d != goat][0]

        if switch_choice == car:
            success += 1

    return success / N


print(switch(100000))
```

## This verifies that switching increases odds to $\frac{2}{3}$

This matches the theoretical results. Thus, if you're ever on Monty Hall, you should switch!

## Let's Check the Math

> This is still such a weird result… what's the theory behind it?

Label the three doors `a`, `b`, and `c`. We analyze both strategies: staying and switching.

---

### Case 1: Stay

Without loss of generality, suppose we initially pick door `a`, and the host opens door `c`, revealing a goat. Let `X` be the random variable representing the location of the car.

Initially,

$$
P(X = a) = \frac{1}{3},
$$

since each door is equally likely to hide the car.

After the host opens door `c`, we gain information about that door, but the probability that our original choice `a` contains the car does **not** change. This is because the host is required to open a door with a goat, so this action does not make our initial choice more likely to be correct.

Therefore, if we stay with door `a`, our probability of winning remains

$$
P(\text{win by staying}) = \frac{1}{3}.
$$

---

### Case 2: Switch

Again, suppose we initially pick door `a`, and the host opens door `c`, revealing a goat. We now switch to door `b`.

Initially,

$$
P(X = a) = \frac{1}{3},
$$

so the probability that our original choice is **wrong** is

$$
P(X \neq a) = \frac{2}{3}.
$$

This means the car is behind one of the other two doors (`b` or `c`) with total probability `2/3`.

When the host opens door `c` and shows a goat, one of those two possibilities is eliminated. Because the host never opens the door with the car, the entire remaining probability mass transfers to the only unopened door, `b`.

Therefore,

$$
P(\text{win by switching}) = \frac{2}{3}.
$$

---

**Intuition in one line:**  
Your first guess is wrong `2/3` of the time, and switching corrects that mistake.
